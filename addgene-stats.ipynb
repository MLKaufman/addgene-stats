# Analysis of Brzezinski Lab Plasmid Requests from Addgene

https://www.addgene.org/depositor-collections/brzezinski-lab-crispr-collection/

```python
import pandas as pd
```

```python
df = pd.read_csv('Addgene-Requests-for-Materials-Michael-Kaufman-2005-2023-2023-Mar-08-18-40.csv')
```

```python
print("Total number of plasmids requested: ", 
      df['Material'].value_counts().sum())
```

```python
df
```

```python
(
pd.to_datetime(df['Date Ordered'])
.value_counts()
.plot(title='Addgene Plasmid Requests by Date', figsize=(15, 5))
.set_ylabel('Request Count')
);
```

```python
(
df['Material']
.value_counts()
.sort_values()
.plot(kind='barh', title='Addgene Plasmid Requests')
.set_xlabel('Request Count')
);
```

```python
(
df['Material']
.value_counts()
.plot(kind='pie', title='Addgene Plasmid Requests', autopct='%1.1f%%', figsize=(6, 6))
.set_ylabel('')
);
```
